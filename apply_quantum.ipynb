```python
test = test.merge(gp, on=['param_1','param_2','category_name'],how='left')
```

```python
tmp_test = test[['item_id','deal_probability','exact_bins']].copy()
```

```python
tmp_test['deal_probability'] = blend['deal_probability']
```

```python
import numpy as np
from tqdm import tqdm_notebook as tqdm
dp = []
count = 0
for idx, c_row in tqdm(tmp_test.iterrows(), total= len(tmp_test)):
    
    tmp_dp = c_row['deal_probability']
    if(str(c_row['exact_bins'])== str(np.nan)):
        dp.append(tmp_dp)
    else:
        
        count += 1
        if(c_row['exact_bins']==1):
            dp.append(0.0)
        else:
            total_set = np.linspace(0.0,1.0, num=c_row['exact_bins'])
            for jdx,ts in enumerate(total_set):
                if(tmp_dp > ts):
                    before = jdx
                    after  =  jdx+1
            print("before:{}  after:{} actual:{}  total_set {}".format(before, after,tmp_dp, total_set))
            hoho = np.array([tmp_dp-total_set[before],total_set[after]-tmp_dp])
            index = np.argmin(hoho)
            if(index==0):
                dp.append(total_set[before])
            else:
                dp.append(total_set[after])
print(count)
```

```python
tmp_test['deal_probability']=dp
```
